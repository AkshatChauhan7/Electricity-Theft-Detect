```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

np.random.seed(42)

# =====================================
# 1. LOAD DATA
# =====================================
consumer_df = pd.read_csv("patiala_consumer_6month.csv")
feeder_df = pd.read_csv("patiala_feeder_6month.csv")

consumer_df['timestamp'] = pd.to_datetime(consumer_df['timestamp'])
feeder_df['timestamp'] = pd.to_datetime(feeder_df['timestamp'])

# =====================================
# 2. AGGREGATE CONSUMER → FEEDER
# =====================================
agg = consumer_df.groupby(['timestamp','feeder_id'])['recorded_consumption_kWh'].sum().reset_index()
agg.rename(columns={'recorded_consumption_kWh':'total_meter_load_kWh'}, inplace=True)

# =====================================
# 3. MERGE
# =====================================
df = pd.merge(feeder_df, agg, on=['timestamp','feeder_id'])
df = df.sort_values(['feeder_id','timestamp']).reset_index(drop=True)

# =====================================
# 4. ANOMALY INJECTION (CRITICAL FIX)
# =====================================
def inject_anomalies(row):
    r = np.random.rand()

    if r < 0.10:
        # THEFT: transformer higher than meter
        row['transformer_load_kWh'] *= np.random.uniform(1.15, 1.35)
        row['status'] = "THEFT"

    elif r < 0.15:
        # FAULT: irregular behavior
        row['transformer_load_kWh'] *= np.random.uniform(0.5, 1.5)
        row['status'] = "FAULT"

    else:
        row['status'] = "NORMAL"

    return row

df = df.apply(inject_anomalies, axis=1)

# =====================================
# 5. FEATURE ENGINEERING
# =====================================

df['loss_kWh'] = df['transformer_load_kWh'] - df['total_meter_load_kWh']
df['loss_pct'] = df['loss_kWh'] / df['transformer_load_kWh']

window = 3

df['rolling_avg_loss'] = df.groupby('feeder_id')['loss_pct'].transform(lambda x: x.rolling(window,1).mean())
df['rolling_std_loss'] = df.groupby('feeder_id')['loss_pct'].transform(lambda x: x.rolling(window,1).std())
df['rolling_max_loss'] = df.groupby('feeder_id')['loss_pct'].transform(lambda x: x.rolling(window,1).max())

tau = 0.15
df['high_loss_flag'] = (df['loss_pct'] > tau).astype(int)

def consecutive(series):
    count = 0
    result = []
    for v in series:
        if v == 1:
            count += 1
        else:
            count = 0
        result.append(count)
    return result

df['consecutive_high_loss_count'] = df.groupby('feeder_id')['high_loss_flag'].transform(consecutive)

df['loss_diff'] = df.groupby('feeder_id')['loss_pct'].diff()
df['hour'] = df['timestamp'].dt.hour

df = df.dropna().reset_index(drop=True)

# =====================================
# 6. WINDOW DATASET
# =====================================
window_size = 3
rows = []

for fid in df['feeder_id'].unique():
    sub = df[df['feeder_id']==fid].reset_index(drop=True)

    for i in range(len(sub)-window_size+1):
        w = sub.iloc[i:i+window_size]

        rows.append({
            'feeder_id': fid,
            'start_time': w['timestamp'].iloc[0],
            'avg_loss_pct': w['loss_pct'].mean(),
            'max_loss_pct': w['loss_pct'].max(),
            'std_loss_pct': w['loss_pct'].std(),
            'high_loss_count': w['high_loss_flag'].sum(),
            'max_consecutive_high_loss': w['consecutive_high_loss_count'].max(),
            'avg_loss_diff': w['loss_diff'].mean(),
            'label': w['status'].mode()[0]
        })

window_df = pd.DataFrame(rows)

# =====================================
# 7. ADVANCED FEATURES
# =====================================
window_df['peak_deviation'] = window_df['max_loss_pct'] - window_df['avg_loss_pct']
window_df['loss_consistency'] = window_df['std_loss_pct'] / (window_df['avg_loss_pct'] + 1e-5)

window_df['rolling_baseline'] = window_df.groupby('feeder_id')['avg_loss_pct'].transform(
    lambda x: x.rolling(window=10, min_periods=1).mean()
)

window_df['relative_loss'] = window_df['avg_loss_pct'] / (window_df['rolling_baseline'] + 1e-5)

# =====================================
# 8. TRAIN TEST SPLIT
# =====================================
window_df = window_df.sort_values('start_time').reset_index(drop=True)

split_index = int(0.7 * len(window_df))
train_df = window_df.iloc[:split_index].copy()
test_df = window_df.iloc[split_index:].copy()

# =====================================
# 9. FEATURES
# =====================================
features = [
    'avg_loss_pct',
    'max_loss_pct',
    'std_loss_pct',
    'high_loss_count',
    'max_consecutive_high_loss',
    'avg_loss_diff',
    'peak_deviation',
    'loss_consistency',
    'relative_loss'
]

X_train = train_df[features]
X_test = test_df[features]

# =====================================
# 10. SCALING
# =====================================
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# =====================================
# 11. TRAIN MODEL
# =====================================
train_normal_idx = train_df['label'] == "NORMAL"

model = IsolationForest(
    n_estimators=130,
    contamination=0.12,
    random_state=42
)

model.fit(X_train_scaled[train_normal_idx])

# =====================================
# 12. PREDICT
# =====================================
test_df = test_df.copy()

test_df.loc[:, 'anomaly_raw'] = model.predict(X_test_scaled)
test_df.loc[:, 'anomaly_raw'] = test_df['anomaly_raw'].map({1:0, -1:1})

test_df.loc[:, 'anomaly_score'] = -model.decision_function(X_test_scaled)

# =====================================
# 13. FINAL DECISION
# =====================================
threshold = np.quantile(-model.decision_function(X_train_scaled), 0.90)

def final_decision(row):
    if row['anomaly_score'] > threshold:
        return "ANOMALY"
    else:
        return "NORMAL"

test_df.loc[:, 'final_status'] = test_df.apply(final_decision, axis=1)

# =====================================
# 14. EVALUATION
# =====================================
test_df.loc[:, 'true_binary'] = test_df['label'].apply(lambda x: 0 if x=="NORMAL" else 1)
test_df.loc[:, 'pred_binary'] = test_df['final_status'].apply(lambda x: 0 if x=="NORMAL" else 1)

print("\n🔥 FINAL PERFORMANCE 🔥")
print("Accuracy:", accuracy_score(test_df['true_binary'], test_df['pred_binary']))
print(classification_report(test_df['true_binary'], test_df['pred_binary']))
print(confusion_matrix(test_df['true_binary'], test_df['pred_binary']))

# =====================================
# 15. CHECK DISTRIBUTION
# =====================================
print("\nLabel Distribution:")
print(train_df['label'].value_counts())
print(test_df['label'].value_counts())
```
