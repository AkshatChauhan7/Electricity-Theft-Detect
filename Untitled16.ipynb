```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# =====================================
# 1. LOAD DATA
# =====================================
df = pd.read_csv("window_6month.csv")
df['start_time'] = pd.to_datetime(df['start_time'])

# =====================================
# 2. CLEAN DATA
# =====================================
df = df.replace([np.inf, -np.inf], np.nan)
df = df.dropna().reset_index(drop=True)

# =====================================
# 3. FEATURE ENGINEERING (ADVANCED)
# =====================================

# New strong features
df['peak_deviation'] = df['max_loss_pct'] - df['avg_loss_pct']
df['loss_consistency'] = df['std_loss_pct'] / (df['avg_loss_pct'] + 1e-5)

df['rolling_baseline'] = df.groupby('feeder_id')['avg_loss_pct'].transform(
    lambda x: x.rolling(window=10, min_periods=1).mean()
)

df['relative_loss'] = df['avg_loss_pct'] / (df['rolling_baseline'] + 1e-5)

# =====================================
# 4. FEATURE SELECTION
# =====================================
features = [
    'avg_loss_pct',
    'max_loss_pct',
    'std_loss_pct',
    'high_loss_count',
    'max_consecutive_high_loss',
    'avg_loss_diff',
    'peak_deviation',
    'loss_consistency',
    'relative_loss'
]

# =====================================
# 5. TIME-BASED SPLIT (IMPORTANT)
# =====================================
df = df.sort_values('start_time').reset_index(drop=True)

split_index = int(0.7 * len(df))

train_df = df.iloc[:split_index].copy()
test_df = df.iloc[split_index:].copy()

# =====================================
# 6. SCALING (FIT ONLY ON TRAIN)
# =====================================
scaler = StandardScaler()

X_train = train_df[features]
X_test = test_df[features]

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# =====================================
# 7. TRAIN MODEL (ONLY NORMAL DATA)
# =====================================
train_normal_idx = train_df['label'] == 'NORMAL'

model = IsolationForest(
    n_estimators=130,
    contamination=0.11,
    max_samples='auto',
    random_state=42
)

model.fit(X_train_scaled[train_normal_idx])

# =====================================
# 8. PREDICT ON TEST DATA
# =====================================
test_df['anomaly_raw'] = model.predict(X_test_scaled)
test_df['anomaly_raw'] = test_df['anomaly_raw'].map({1: 0, -1: 1})

test_df['anomaly_score'] = -model.decision_function(X_test_scaled)

# =====================================
# 9. RULE-BASED FILTER
# =====================================
LOSS_THRESHOLD = 0.18

test_df['rule_flag'] = (
    (test_df['avg_loss_pct'] > LOSS_THRESHOLD) &
    (test_df['high_loss_count'] >= 2)
).astype(int)

# =====================================
# 10. SCORE THRESHOLD (FROM TRAIN DATA)
# =====================================
train_scores = -model.decision_function(X_train_scaled)
SCORE_THRESHOLD = np.quantile(train_scores, 0.90)

# =====================================
# 11. FINAL DECISION
# =====================================
def final_decision(row):
    if row['rule_flag'] == 1:
        return "ANOMALY"
    elif row['anomaly_score'] > SCORE_THRESHOLD and row['anomaly_raw'] == 1:
        return "ANOMALY"
    else:
        return "NORMAL"

test_df['final_status'] = test_df.apply(final_decision, axis=1)

# =====================================
# 12. EVALUATION (ONLY TEST DATA)
# =====================================
test_df['true_binary'] = test_df['label'].apply(lambda x: 0 if x == "NORMAL" else 1)
test_df['pred_binary'] = test_df['final_status'].apply(lambda x: 0 if x == "NORMAL" else 1)

accuracy = accuracy_score(test_df['true_binary'], test_df['pred_binary'])

print("\n🔥 TEST SET PERFORMANCE 🔥")
print(f"\n✅ Accuracy: {accuracy:.4f}")

print("\n=== Classification Report ===")
print(classification_report(test_df['true_binary'], test_df['pred_binary']))

cm = confusion_matrix(test_df['true_binary'], test_df['pred_binary'])
print("\n=== Confusion Matrix ===")
print(cm)

tn, fp, fn, tp = cm.ravel()

print("\nDetailed Breakdown:")
print(f"True Positives: {tp}")
print(f"False Positives: {fp}")
print(f"False Negatives: {fn}")
print(f"True Negatives: {tn}")

# =====================================
# 13. SAVE OUTPUT
# =====================================
test_df.to_csv("final_test_predictions.csv", index=False)

print("\n✅ Test predictions saved as final_test_predictions.csv")
```
